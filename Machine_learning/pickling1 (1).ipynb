```python
import pandas as pd 
df = pd.read_csv('covid_19_data.csv')
```

```python
print(df)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Province/State','Country/Region']
le = LabelEncoder()
for c in cols:
    df[c]=le.fit_transform(df[c])
df.head()
```

```python
# Convert all values in 'Province/State' to strings
df['ObservationDate'] = df['ObservationDate'].astype(str)

# Now apply LabelEncoder
le = LabelEncoder()
df['ObservationDate'] = le.fit_transform(df['ObservationDate'])
df
```

```python
df['Last Update'] = df['Last Update'].astype(str)

# Now apply LabelEncoder
le = LabelEncoder()
df['Last Update'] = le.fit_transform(df['Last Update'])
df
```

```python
# Convert all values in 'Province/State' to strings
df['Province/State'] = df['Province/State'].astype(str)

# Now apply LabelEncoder
le = LabelEncoder()
df['Province/State'] = le.fit_transform(df['Province/State'])
df
```

```python
X=df.drop(['ObservationDate'],axis=1)
X.head()
```

```python
Y=df[['ObservationDate']]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sl = StandardScaler()
X_train=sl.fit_transform(X_train)
X_test= sl.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```
